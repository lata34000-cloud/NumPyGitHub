## Numpy function
* **Array Creation Functions:-** (array,zeros,ones,eye,full)
* **Mathematical Functions:-** (sum,mean,max,min,sqrt)
* **Shape & Dimension Functions:-** (ndim,shape,size)
* **Random Functions:-** (random.rand,random.randint,random.randn)
* **Sequence Functions:-** (arange,linspace)
* **Matrix Functions:-** (dot,transpose)


### Array Creation Functions :-

**Array:-** *An array is a data structure that stores multiple values of the same data type in a single variable.*

```python
a = np.array([1, 2, 3, 4])
a
```

**Zeros:-** *np.Zeros() is a function that creates an array filled with 0 values.*

```python
a=np.zeros((3,3))
a
```

**Ones:-** *np.Ones() is a function that created an array filled with values 1.*

```python
a=np.ones((2,4))
a
```

**eye:-** *The Eye() function in NumPy is used to create an identity matrix, where the diagonal elements are 1 and all other elements are 0.*

```python
a=np.eye(3)
a
```

**Full:-** *The full() function is used to create an array filled with a specified value.*

```python
arr=np.full((3,3),7)
arr
```

### Mathematical Functions :-

**sum:-** *sum() is a function used to add all the elements of an array.*

```python
arr = np.array([10, 20, 30, 40])
print(arr.sum())
```

**mean:-** *mean() is used to calculate the average of element.*

```python
arr = np.array([1, 2, 3, 4])
print(arr.mean())
```

**max:-** *max() is a function used to find the largest value in an array.*

```python
arr = np.array([10, 25, 7, 40, 18])
print(arr.max())
```

**min:-** *min() is a function used to find the smallest value in an array.*

```python
arr = np.array([10, 25, 7, 40, 18])
print(arr.min())
```

**sqrt:-** *sqrt() is a function used to find the square root of each element in an array.*

```python
arr = np.array([4,9,16,25])
print(np.sqrt(arr))
```

### Shape & Dimension Functions :-

**ndim:-** *ndim is used to find the number of dimensions of an array.*

```python
arr = np.array([10, 20, 30,40])
print(arr.ndim)
```

**shape:-** *shape is used to find the number of rows and columns (dimensions) of an array.*

```python
arr = np.array([[1,2,3],[4,5,6]])
print(arr.shape)
```

**size:-** *size is used to find the total number of elements in an array.*

```python
arr = np.array([10, 20, 30, 40,50,60])
print(arr.size)
```

### Random Functions :-

**random.rand:-** *random.rand() is used to create an array of random numbers between 0 and 1.*

```python
arr = np.random.rand(3,3)
print(arr)
```

**random.randint:-** *It is random whole number between a start and end value.*

```python
arr=np.random.randint(0,20,(3,3))
print(arr)
```

**random.randn:-** *random.randn() give random number that can negative, Zero, positive.*

```python
arr=np.random.randn(2,3)
print(arr)
```

### Sequence Functions :-

**arange:-** *arange() is used to create a sequence of numbers within a specified range.*

```python
arr=np.arange(0,20,2)
print(arr)
```

**Linspace:-** *linspace() is used to create a sequence of evenly spaced numbers between a start and an end value.*

```python
arr=np.linspace(0,10,5)
print(arr)
```

### Matrix Functions :-

**Dot:-** *dot() is used to perform matrix multiplication between two arrays.*

```python
a = np.array([1, 2, 3,4,5])
b = np.array([6,7,8,9,10])
result = np.dot(a, b)
print(result)
```

**Transpose:-** *transpose() is used to swap rows and columns of an array.*

```python
arr=np.array([[1,2,3],[4,5,6]])
print(arr.T)
```

### Other Functions :-

**argmax:-** *argmax() is used to find the index of the maximum value in an array.*

```python
arr = np.array([1,2,3,4,5])
print(np.argmax(arr))
```

**agrmin:-** *argmin() is used to find the index of the minimum value in an array.*

```python
arr = np.array([10, 45, 23, 5, 34])
print(np.argmin(arr))
```

**std:-** *std() is used to calculate the standard deviation of elements in an array, which shows how much the values deviate from the mean.*

```python
arr = np.array([10, 20, 30, 40, 50])
print(np.std(arr))
```

**flatten:-** *flatten() is used to convert a multi-dimensional array into a one-dimensional (1D) array.*

```python
arr = np.array([[1, 2, 3],[4, 5, 6]])
print(arr.flatten())
```

**ravel:-** *ravel() is used to convert any array into a 1D(faltend) array.*

```python
arr = np.array([[1, 2, 3],[4, 5, 6]])
print(arr.ravel())
```

**reshape:-** *reshape() is used to change the shape of an array without changing its data.*

```python
arr = np.array([[1,2,3],[4,5,6]])
print(arr.reshape(2,3))
```

**diag:-** *diag() is used to extract the diagonal elements of a matrix or create a diagonal matrix from a 1D array.*

```python
arr = np.array([1, 2, 3, 4])
print(np.diag(arr))
```

### Arithmetic Operations

```python
a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])
print(a + b)

print(a - b)

print(a * b)

print(a / b)
```

### Sort Array

```python
arr = np.array([40, 10, 30, 20])
print(np.sort(arr))
```

## Difference Between NumPy Array and Python List

| Basis | NumPy Array | Python List |
|------|-------------|-------------|
| Data Type | Same type of elements | Different types of elements |
| Speed | Faster | Slower |
| Memory | Uses less memory | Uses more memory |
| Mathematical Operations | Directly supported | Need loops |
| Performance | Best for large data | Better for small data |
| Dimension | Supports multi-dimensional | Mostly one-dimensional |
| Library | NumPy library | Python built-in |


```python
list1 = [1, 2, 3, 4]
array1 = np.array([1,2,3,4])
print("List Addition: ",list1 + list1)
print("Numpy array Addition : ", array1 +array1)
```
